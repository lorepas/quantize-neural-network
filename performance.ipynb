```python
#svhn 1-2-4
f = open("train_log/svhn-dorefa-1,2,4/stats_def_1.json","r")

data = json.load(f)
accuracy_1 = []
val_accuracy_1 = []
for ob in data:
  accuracy_1.append(ob["accuracy"])
  val_accuracy_1.append(ob["val_accuracy"])
```

```python
#mnist 4-6-16
f = open("train_log/mnist-dorefa-4,6,16/stats_def_first&last_avg_pooling_4616.json","r")

data = json.load(f)
accuracy_8 = []
val_accuracy_8 = []
for ob in data:
  accuracy_8.append(ob["accuracy"])
  val_accuracy_8.append(ob["val_accuracy"])
```

```python
# create figure
fig, (acc_svhn,acc_mnist) = plt.subplots(1,2, figsize=(15,6))

# range for x axis
epochs = range(len(accuracy_1))

# plot accuracy curves
# acc_svhn.plot(epochs, accuracy_1, color='tab:red', marker='.', linestyle='-', label='1-2-4 training', linewidth=1)
# acc_svhn.plot(epochs, val_accuracy_1, color='tab:blue', marker='.', linestyle='-', label='1-2-4 testing', linewidth=1)
# acc_svhn.plot(epochs, accuracy_2, color='tab:green', marker='*', linestyle='-', label='1-2-4_flq training', linewidth=1)
# acc_svhn.plot(epochs, val_accuracy_2, color='tab:orange', marker='*', linestyle='-', label='1-2-4_flq testing', linewidth=1)
acc_svhn.plot(epochs, accuracy_3, color='tab:red', marker='o', linestyle='-', label='1-2-4_flq_avg training', linewidth=1)
acc_svhn.plot(epochs, val_accuracy_3, color='tab:blue', marker='o', linestyle='-', label='1-2-4_flq_avg testing', linewidth=1)
acc_svhn.plot(epochs, accuracy_4, color='tab:green', marker='*', linestyle='-', label='6-6-6 training', linewidth=1)
acc_svhn.plot(epochs, val_accuracy_4, color='tab:orange', marker='*', linestyle='-', label='6-6-6 testing', linewidth=1)
acc_svhn.plot(epochs, accuracy_5, color='tab:brown', marker='^', linestyle='-', label='4-6-16 training', linewidth=1)
acc_svhn.plot(epochs, val_accuracy_5, color='tab:purple', marker='^', linestyle='-', label='4-6-16 testing', linewidth=1)

# plot loss curves
acc_mnist.plot(epochs, accuracy_6, color='tab:red', marker='o', linestyle='-', label='1-2-4_flq_avg training', linewidth=1)
acc_mnist.plot(epochs, val_accuracy_6, color='tab:blue', marker='o', linestyle='-', label='1-2-4_flq_avg testing', linewidth=1)
acc_mnist.plot(epochs, accuracy_7, color='tab:green', marker='*', linestyle='-', label='6-6-6 training', linewidth=1)
acc_mnist.plot(epochs, val_accuracy_7, color='tab:orange', marker='*', linestyle='-', label='6-6-6 testing', linewidth=1)
acc_mnist.plot(epochs, accuracy_8, color='tab:brown', marker='^', linestyle='-', label='4-6-16 training', linewidth=1)
acc_mnist.plot(epochs, val_accuracy_8, color='tab:purple', marker='^', linestyle='-', label='4-6-16 testing', linewidth=1)

# accuracy graph style options
acc_svhn.grid(axis='y', linestyle='--', linewidth=0.5)
acc_svhn.set_xlabel('Epochs')
acc_svhn.set_ylabel('Accuracy')
acc_svhn.set_title('Accuracy SVHN dataset', fontdict={'fontsize':18}, pad=20)
acc_svhn.legend(loc='lower right')

# loss graph style options
acc_mnist.grid(axis='y', linestyle='--', linewidth=0.5)
acc_mnist.set_xlabel('Epochs')
acc_mnist.set_ylabel('Accuracy')
acc_mnist.set_title('Accuracy MNIST dataset', fontdict={'fontsize':18}, pad=20)
acc_mnist.legend(loc='center right')

# layout adjustment
plt.tight_layout()
```

```python
ep = [i+1 for i in epochs]
table_acc = {"Epochs" : ep, "Accuracy 1-2-4 svhn":accuracy_3, "Accuracy 6-6-6 svhn":accuracy_4, 
             "Accuracy 4-6-16 svhn":accuracy_5, "Accuracy 1-2-4 mnist":accuracy_6, "Accuracy 6-6-6 mnist":accuracy_7, 
             "Accuracy 4-6-16 mnist":accuracy_8}
table_val_acc = {"Epochs" : ep, "Val accuracy 1-2-4 svhn":val_accuracy_3, "Val accuracy 6-6-6 svhn":val_accuracy_4, 
             "Val accuracy 4-6-16 svhn":val_accuracy_5, "Val accuracy 1-2-4 mnist":val_accuracy_6, "Val accuracy 6-6-6 mnist":val_accuracy_7, 
             "Val accuracy 4-6-16 mnist":val_accuracy_8}

print("ACCURACY\n")
print(tabulate(table_acc, headers='keys', tablefmt='fancy_grid'))
print("\nVALIDATION ACCURACY\n")
print(tabulate(table_val_acc, headers='keys', tablefmt='fancy_grid'))
```

```python
acc_tab = tabulate(table_acc, headers='keys')
val_acc_tab = tabulate(table_val_acc, headers='keys')

fa = open("accuracy.csv","w")
fa.write(acc_tab)
fa.close()

fva = open("val_accuracy.csv", "w")
fva.write(val_acc_tab)
fva.close()
```
