```python
from google.colab import drive
import json
import matplotlib.pyplot as plt
from tabulate import tabulate
drive.mount('/content/gdrive', force_remount=True)

%cd gdrive/MyDrive/SEAI_Project
```

```python
#mnist 1-2-4
f = open("train_log/mnist-dorefa-1,2,4/stats.json","r")

data = json.load(f)
accuracy_1 = []
val_accuracy_1 = []
for ob in data:
  accuracy_1.append(ob["accuracy"])
  val_accuracy_1.append(ob["val_accuracy"])
```

```python
#mnist 1-4-6
f = open("train_log/mnist-dorefa-1,4,6/stats.json","r")

data = json.load(f)
accuracy_2 = []
val_accuracy_2 = []
for ob in data:
  accuracy_2.append(ob["accuracy"])
  val_accuracy_2.append(ob["val_accuracy"])
```

```python
#mnist 1-6-2
f = open("train_log/mnist-dorefa-1,6,2/stats.json","r")

data = json.load(f)
accuracy_3 = []
val_accuracy_3 = []
for ob in data:
  accuracy_3.append(ob["accuracy"])
  val_accuracy_3.append(ob["val_accuracy"])
```

```python
#mnist 1-6-4

f = open("train_log/mnist-dorefa-1,6,4/stats.json","r")

data = json.load(f)
accuracy_4 = []
val_accuracy_4 = []
for ob in data:
  accuracy_4.append(ob["accuracy"])
  val_accuracy_4.append(ob["val_accuracy"])
```

```python
#mnist 4-4-6
f = open("train_log/mnist-dorefa-4,4,6/stats.json","r")

data = json.load(f)
accuracy_5 = []
val_accuracy_5 = []
for ob in data:
  accuracy_5.append(ob["accuracy"])
  val_accuracy_5.append(ob["val_accuracy"])
```

```python
#mnist 6-4-4
f = open("train_log/mnist-dorefa-6,4,4/stats.json","r")

data = json.load(f)
accuracy_6 = []
val_accuracy_6 = []
for ob in data:
  accuracy_6.append(ob["accuracy"])
  val_accuracy_6.append(ob["val_accuracy"])
```

```python
# create figure
fig, (acc_mnist,acc_mnist_ok) = plt.subplots(1,2, figsize=(15,6))

# range for x axis
epochs = range(len(accuracy_1))

# plot accuracy curves
acc_mnist.plot(epochs, accuracy_1, color='tab:red', marker='o', linestyle='-', label='1-2-4 training', linewidth=1)
acc_mnist.plot(epochs, val_accuracy_1, color='tab:blue', marker='o', linestyle='-', label='1-2-4 testing', linewidth=1)
acc_mnist.plot(epochs, accuracy_2, color='tab:cyan', marker='s', linestyle='-', label='1-4-6 training', linewidth=1)
acc_mnist.plot(epochs, val_accuracy_2, color='tab:gray', marker='s', linestyle='-', label='1-4-6 testing', linewidth=1)
acc_mnist.plot(epochs, accuracy_5, color='tab:brown', marker='^', linestyle='-', label='4-4-6 training', linewidth=1)
acc_mnist.plot(epochs, val_accuracy_5, color='tab:purple', marker='^', linestyle='-', label='4-4-6 testing', linewidth=1)
acc_mnist.plot(epochs, accuracy_6, color='tab:green', marker='*', linestyle='-', label='6-4-4 training', linewidth=1)
acc_mnist.plot(epochs, val_accuracy_6, color='tab:orange', marker='*', linestyle='-', label='6-4-4 testing', linewidth=1)

# plot loss curves
acc_mnist_ok.plot(epochs, accuracy_3, color='tab:red', marker='o', linestyle='-', label='1-6-2 training', linewidth=1)
acc_mnist_ok.plot(epochs, val_accuracy_3, color='tab:blue', marker='o', linestyle='-', label='1-6-2 testing', linewidth=1)
acc_mnist_ok.plot(epochs, accuracy_4, color='tab:green', marker='*', linestyle='-', label='1-6-4 training', linewidth=1)
acc_mnist_ok.plot(epochs, val_accuracy_4, color='tab:orange', marker='*', linestyle='-', label='1-6-4 testing', linewidth=1)

# accuracy graph style options
acc_mnist.grid(axis='y', linestyle='--', linewidth=0.5)
acc_mnist.set_xlabel('Epochs')
acc_mnist.set_ylabel('Accuracy')
acc_mnist.set_title('Accuracy MNIST bad prediction', fontdict={'fontsize':18}, pad=20)
acc_mnist.legend(loc='lower right')

# loss graph style options
acc_mnist_ok.grid(axis='y', linestyle='--', linewidth=0.5)
acc_mnist_ok.set_xlabel('Epochs')
acc_mnist_ok.set_ylabel('Accuracy')
acc_mnist_ok.set_title('Accuracy MNIST good prediction', fontdict={'fontsize':18}, pad=20)
acc_mnist_ok.legend(loc='lower right')

# layout adjustment
plt.tight_layout()
```

```python
ep = [i+1 for i in epochs]
table_acc = {"Epochs" : ep, "1-2-4":accuracy_1, "1-4-6":accuracy_2, 
             "4-4-6":accuracy_5, "6-4-4":accuracy_6, "1-6-2":accuracy_3, "1-6-4":accuracy_4}
table_val_acc = {"Epochs" : ep, "1-2-4":val_accuracy_1, "1-4-6":val_accuracy_2, 
             "4-4-6":val_accuracy_5, "6-4-4":val_accuracy_6, "1-6-2":val_accuracy_3, "1-6-4":val_accuracy_4}

print("TRAINING ACCURACY|")
print("------------------\n")
print(tabulate(table_acc, headers='keys', tablefmt='github'))
print("\n--------------")
print("TEST ACCURACY|")
print("--------------\n")
print(tabulate(table_val_acc, headers='keys', tablefmt='github'))
```

```python
acc_tab = tabulate(table_acc, headers='keys')
val_acc_tab = tabulate(table_val_acc, headers='keys')

fa = open("appr_accuracy.csv","w")
fa.write(acc_tab)
fa.close()

fva = open("appr_val_accuracy.csv", "w")
fva.write(val_acc_tab)
fva.close()
```
